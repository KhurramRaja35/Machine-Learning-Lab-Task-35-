```python
# we are now ploting through functional programing , there is also alternate to plot graphs using object oriented way
```

```python
fig = plt.figure(figsize = (12,4))
axis = fig.add_axes([0.1,0.1,0.8,0.8])
axis.plot(x,y)
```

```python
fig1 = plt.figure()
axis1 = fig1.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig1.add_axes([0.2,0.5,0.2,0.3])
axis1.plot(x,y)
axis2.plot(y,x)
axis1.set_xlabel('X')
axis1.set_ylabel('Y')
axis1.set_title('Title')
```

```python
fig , axis = plt.subplots(1,2)
axis[0].plot(x,y)
axis[1].plot(y,x)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, x**2, label = "x**2",color = 'green', linewidth = 3, marker = 'o' , markersize=10, markerfacecolor = 'yellow')
axes.plot(x,x**3, label = "x**3")
axes.legend(loc=0)
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.displot(tips['total_bill'])
# displot ---> distribution plot for univariant data....
```

```python
sns.jointplot(x = 'total_bill' , y ='tip' , data= tips , kind = 'hex' )
# jointplot for bivariant data
```

```python
sns.pairplot(data=tips , hue='smoker')
# pairplot data k har feature ka dosry k sth pair kr k plot kr de ga..
```

```python
sns.rugplot(data=tips['total_bill'])
```

```python
sns.barplot(data = tips , x = 'sex' , y = 'total_bill' , palette = 'hot' , hue = 'smoker' , estimator = 'sum')
```

```python
sns.countplot(data=tips , x = "smoker" , hue = 'smoker' , palette = 'hot')
```

```python
 flierprops = {'marker' : 'o', 'markerfacecolor' : 'red' , 'markeredgecolor' : 'yellow' ,'markersize' : '8'}
sns.boxplot(data = tips , x='total_bill' , y = 'day' , hue = 'day' , palette = 'hot', flierprops = flierprops)
```
